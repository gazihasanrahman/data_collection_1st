```python
import json
import pandas as pd
import os
import glob
```

```python
location_data_path = 'downloads'

list_of_files = glob.glob(os.path.join(location_data_path, '*.json'))
list_of_files
```

```python
def quick_analysis(file_index):
    file = list_of_files[file_index]
    with open(file, 'r') as f:
        data = json.load(f)
    timestamp = data.get('header').get('timestamp')
    timestamp = pd.to_datetime(timestamp)
    entries = data.get('fixtures')[0].get('races')[0].get('entries')
    entries_df = pd.DataFrame(entries)
    entries_df = entries_df.drop(columns=['favPos', 'favJoint', 'finalPosition', 'jockey', 'weight'])
    entries_expanded = pd.concat([
        entries_df.drop(columns=['equipment', 'startingPrice']),
        entries_df['equipment'].apply(pd.Series).add_prefix('equipment_'),
        entries_df['startingPrice'].apply(pd.Series).add_prefix('startingPrice_')
    ], axis=1)
    entries_expanded['timestamp'] = timestamp
    return entries_expanded
```

```python
quick_analysis(0)
```

```python
list_of_df = []
for i in range(len(list_of_files)):
    result = quick_analysis(i)
    list_of_df.append(result)

# concatenate all dataframes in list_of_df
all_data = pd.concat(list_of_df, ignore_index=True)
all_data


```
